```python
import numpy as np
import matplotlib.pyplot as plt
import mglearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# 데이터셋 
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
print(boston.keys())
```

```python
print(boston.items())
```

```python
import pandas as pd
```

```python
# df = pd.DataFrame({'data':, 'target':, 'feature_names':, 'DESCR':, 'filename':})

df = pd.DataFrame(boston.data, columns=boston.feature_names)
df
```

```python
X = boston.data
y = boston.target
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=100)
```

```python
# 모델 학습시키기 - fit
model = LinearRegression().fit(X_train, y_train)
```

```python
# 모델 예측하기 - predict
pred = model.predict(X_test)
pred
```

```python
y_test
```

```python
# MAE = Prediction = y_test
```

```python
df = pd.DataFrame({'y_test': y_test, 'prediction': pred, "difference": y_test -pred})
```

```python
df
```

```python
k=abs(y_test - pred)
t = sum(k)
```

```python
t
```

```python
t/len(k)
```

```python
#OR

sum(abs(df['difference']))/len(df['difference'])
```

```python
#MSE
```

```python
sum(abs(df['difference']**2))/len(df['difference'])
```

```python
from sklearn.metrics import mean_squared_error
```

```python
p = mean_squared_error(y_test, pred)
```

```python
#RMSE
q = np.sqrt(p)
q
```

```python
# 훈련 데이터셋 및 테스트 데이터셋 점수 확인
print('훈련 데이터 셋 점수  : {:.2f}'.format(model.score(X_train, y_train)))
print('테스트 데이터 셋 점수  : {:.2f}'.format(model.score(X_test, y_test)))
```
