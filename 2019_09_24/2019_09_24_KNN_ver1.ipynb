```python
from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = cancer.data
y = cancer.target

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = cancer.target, random_state=60)
##stratify can only applied to 범주형, NOT 연속형
```

```python
# 머신러닝 모델 만들기 & 예측하기
# (1) 모델 만들기
# (2) 모델 학습시키기 model.fit
# (3) 모델 예측하기 model.predict
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)
pred = model.predict(X_test)
pred
```

```python
#얼마나 맞추었나 accuracy 추측하기
(pred == y_test).sum()/len(pred)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'y_test': y_test, 'prediction': pred})
df
```

```python
acc_tr = model.score(X_train, y_train)
acc_test = model.score(X_test, y_test)
print('훈련 셋의 정확도 : {:.2f}'.format(acc_tr))
print('테스트 셋의 정확도 : {:.2f}'.format(acc_test))
```

```python
# for문을 돌려보쟈

training_accuracy = []
test_accuracy = []
k_num = range(1,21)

for n in k_num:
    model = KNeighborsClassifier(n_neighbors=n)
    model.fit(X_train, y_train)
    
    acc_tr = model.score(X_train, y_train)
    acc_test = model.score(X_test, y_test)
    
    training_accuracy.append(acc_tr)
    test_accuracy.append(acc_test)
    
    print("K: {}, tr_acc : {:.2f}, test_acc : {:.2f}".format(n, acc_tr, acc_test))
```
